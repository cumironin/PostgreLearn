```python
# Using -- = -- as equal, in movie_lang column must only english as language
SELECT * FROM movies
WHERE movie_lang = 'English'



# Two of filtering keyword must be right  -- English and 18 --
SELECT * FROM movies
	WHERE movie_lang = 'English'
	AND age_certificate  = '18'
    
# Two of filtering keyword only option, if theres English show it, If there's Chinese show it
# The result of this query is two of row fill --- English and Chinese --- is showed
SELECT * FROM movies
	WHERE movie_lang = 'English'
	OR movie_lang = 'Chinese'

# Filtering movies -- English or Chinese -- means that two of row in column -- movie_lang -- must be shown
# AND is use to force that English or Chinese is shown but must be age to 12
# using ( ) is important in here to push logical sorting
SELECT * FROM movies
WHERE 
	(movie_lang  = 'English'
	OR movie_lang = 'Chinese')
	AND age_certificate = '12'
ORDER BY movie_lang 




# Logical Operator
# Use : > < >= <= = !=
SELECT 
* 
FROM movies
WHERE 
	release_date > '1999-12-31'

# Pakai abjad di -- English --, jadi yang di filter abjadnya
SELECT 
* 
FROM movies
WHERE 
	release_date > 'English'
    
    
    
    
# Limit adalah upaya untuk membatasi row pada table, 
# misal kita cuma mau ngambil director orang america dengan tanggal lahir belakang dan batasi cuma 5 row
SELECT * FROM directors
WHERE nationality = 'American'
ORDER BY date_of_birth DESC 
LIMIT 5

# Offset adalah upaya untuk memulai limit
# Misal kita sudah LIMIT row menjadi 5, maka OFFSET adalah memulai dengan row ke berapa : misal row yang ke 20
SELECT * FROM movies
ORDER BY movie_id 
LIMIT 5 OFFSET 20

# Dengan cara ini pun bisa kita select semua table row tapi mulai dari row yang ke 20
SELECT * FROM movies
ORDER BY movie_id 
OFFSET 20

# Dengan cara ini mirip limit kita membatasi row dengan 5 row
SELECT * FROM movies
ORDER BY movie_length DESC
FETCH FIRST 5 ROW ONLY




# Return true if value = column movie_lang is match with row IN (...)
# value must match with list
# Misal kita mau filter dit table movies cuma ada 3 bahasa aja berdasarkan list yang ada di column movie_lang
SELECT * FROM movies 
WHERE 
	movie_lang IN ('English', 'Chinese', 'Japanese')
ORDER BY movie_lang 

# cara yang lain ya gini pake OR tapi kurang simple dan hasilnya sama    
SELECT * FROM movies 
WHERE 
	movie_lang = 'English'
	OR movie_lang = 'Chinese'
	OR movie_lang = 'Japanese'
ORDER BY movie_lang 

# Return true if value in column is getting rid of
# Misal kita mau filter di table movies yang ga ada cuma angka 13 atau 10 aja selainya boleh
SELECT * FROM movies 
WHERE 
	director_id NOT IN ('13', '10')

    
    
    
# Filter yang mirip dengan fungsi untuk foreach 
# Jadi ngitungnya dari 1991-1995 (92, 93, 94) kehitung semua, bukan cuma 91 & 95 saja yang ke kefilter
SELECT * FROM actors
WHERE date_of_birth BETWEEN '1991-01-01' AND '1995-12-31'
ORDER BY date_of_birth 

# Yang ini kebalikanya pokoknya angka 100-200 itu di getting rid off, diluar itu semua ke shown di row
SELECT * FROM movies 
WHERE 
	movie_length NOT BETWEEN 100 AND 200
ORDER BY movie_length 

# Pake ini juga sama hasilnya kaya diatas cuma lebih rempong aja
SELECT * FROM movies 
WHERE 
	movie_length >= 100 AND movie_length <= 200
ORDER BY movie_length 




# LIKE itu sebenernya pattern matching yang biasanya diapake untuk search 
# % di pendahuluan kata, diakhiran atau di apit 22nya bisa 
# % itu berfungsi untuk menyocokan kata yang di query
# LIKE also case sensitive
SELECT * FROM actors 
WHERE first_name LIKE '%uk%'
ORDER BY first_name 

# Same function as LIKE but ILIKE is not case sensitive
SELECT * FROM actors 
WHERE first_name ILIKE '%uk%'
ORDER BY first_name 




# Null is checking wheter the value of the columns is Null
# Nanti ini yang muncul di filter yang null semua di column tersebut
SELECT * FROM actors
WHERE date_of_birth IS NULL 
OR first_name IS NULL 

# Ini kebalikanya untuk filter column yang getting rid of NULL
SELECT * FROM movies_revenues
WHERE revenues_domestic IS NOT NULL



# CONCAT adalah upaya untuk menggabungkan column 
SELECT concat(first_name,' ',last_name) AS "Actor Name" FROM actors

# CONCAT untuk setiap elemenya : yang ini contohnya dikasih koma untuk setiap column trus digabung
SELECT concat_ws(',',first_name,last_name, date_of_birth) AS "Actor Name" FROM actors

# Select statemen for expression
# ini untuk mengabungkan kedua kolom dan ngasih space diantaranya 
# -- kol_1 = Hana, kol2 = Hanif --, jadinya : -- Hana Hanif --
SELECT first_name ||' '|| last_name FROM actors;
```
