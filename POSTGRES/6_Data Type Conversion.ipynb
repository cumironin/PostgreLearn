```python
# In postgres Data Conversion Divide into 2 Things
# - Implisit : Data conversion done automatically
# - Explisit : Data conversion is done by 'Conversion Function' e.g : CAST

# No Conversion
# Assignment & Data type is same
SELECT * FROM movies
WHERE movie_id = 1

# Implicit Conversion
# movie_id (data type) = Integer
# movie_id (assigment) = Varchar/Char
# Automatically
SELECT * FROM movies
WHERE movie_id = '1'


# Explisit Conversion 
# From Char/Varchar to Insteger
SELECT * FROM movies
WHERE movie_id = integer '1'

# All above method have same result
movie_id|movie_name        |movie_length|movie_lang|age_certificate|release_date|director_id|
--------+------------------+------------+----------+---------------+------------+-----------+
       1|A Clockwork Orange|         112|English   |18             |  1972-02-02|         13|
    

--------------------------------------------------------------------------------------------------------------------

# USING CAST FOR DATA CONVERSION

# target data type : 
# Boolean, Character(Char, varchar, text), Numeric(Integer, Floating point number), Array, JSON, UUID. 
# Hstore (key value), Temporal Type(date, time, timestamp, timestamptz, interval)
# Special type (network address, geometric address)


# Example :
# From character to number
SELECT CAST ('10' AS INTEGER)
int4|
----+
  10|

# From character to boolean
SELECT CAST ('true' AS BOOLEAN)
bool|
----+
true|

# From character to double precision
SELECT CAST ('14.00909' AS DOUBLE PRECISION)
float8  |
--------+
14.00909|

# From character to timestamp
SELECT '2020-02-20 10:30:25.567'::TIMESTAMP;
timestamp              |
-----------------------+
2020-02-20 10:30:25.567|
        
# From character to interval 
SELECT '10 minute'::INTERVAL 
interval|
--------+
00:10:00|

--------------------------------------------------------------------------------------------------------------------

# Implicit to Explicit Conversion

# convert as numeric with big bytes
SELECT CAST(20 AS bigint) ! AS "result"
result             |
-------------------+
2432902008176640000|

# Tambahni 4 angka dibelakangnya
SELECT round(CAST(10 AS NUMERIC),4) AS "result"
result |
-------+
10.0000|

--------------------------------------------------------------------------------------------------------------------

# Table Data Conversion

CREATE TABLE ratings(
	rating_id SERIAL PRIMARY KEY,
	rating VARCHAR(1) NOT NULL
)

# Strings 
INSERT INTO ratings(rating) VALUES
('A'),
('B'),
('C'),
('D')

# Integer
INSERT INTO ratings(rating) VALUES
(1),
(2),
(3),
(4)

# Become
rating_id|rating|
---------+------+
        1|A     |
        2|B     |
        3|C     |
        4|D     |
        5|1     |
        6|2     |
        7|3     |
        8|4     |


# Change Table that if there's a String in the table, change it to 0
SELECT 
	rating_id,
	CASE
		WHEN rating~E'^\\d+$' THEN
			CAST (rating AS INTEGER)
		ELSE
			0
		END AS rating
FROM 
	ratings

rating_id|rating|
---------+------+
        1|     0|
        2|     0|
        3|     0|
        4|     0|
        5|     1|
        6|     2|
        7|     3|
        8|     4|
```
