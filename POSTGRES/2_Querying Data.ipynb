```python
# Select All Table
# use select column instead
SELECT * FROM movies;

# Don't use this for performance wise, because it wil create bootle neck to get db table 
# (all table content will sum it), use column of table that you need instead
SELECT gender, first_name, last_name FROM actors;

# Alias dipake untuk mengganti nama kolom, fungsinya supaya mudah dibaca saja 
# ketimbang -- first_name, lang -- yang cenderung teknikal nama kolomnya
SELECT first_name AS "First Name" FROM actors;
SELECT movie_name AS "Movie", movie_lang AS "Language" FROM movies;

# Select statemen for expression
# ini untuk mengabungkan kedua kolom dan ngasih space diantaranya 
# -- kol_1 = Hana, kol2 = Hanif --, jadinya : -- Hana Hanif --
SELECT first_name ||' '|| last_name FROM actors;

# Pake  -- AS -- untuk menggani nama kolomnya juga
SELECT first_name ||' '|| last_name AS "Full Name" FROM actors;


# SORTING

# Ini untuk sorting (ngurutin) ASC (dari no awal 1 ke 10) atau DESC (dari no 10 ke 1)
SELECT * FROM movies ORDER BY release_date DESC, movie_name ASC;

# sorting  by last_name with change in it column
SELECT
	first_name, last_name AS surname
FROM actors
ORDER BY last_name;


# ini untuk sort row by expression, nanti diitung berapa row disitu untuk first_name dan dimasukin ke column len
SELECT 
	first_name, 
	LENGTH (first_name) as len 
	FROM actors order by LEN desc 

# di sorting berdasarkan urutan kolom -- firstname : 1, last_name: 2, date_of_birth: 3 --
SELECT first_name, last_name, date_of_birth  FROM actors
ORDER BY 
	1 ASC, 
	3 DESC 


# Sorting num (null value) into first row
SELECT * FROM demo_sorting ORDER BY num DESC 


# Pake DISTINC untuk mensorting duplicate row di column yang di select
# Misal banyak bahasa yang duplicate di column tersebut maka akan di ambil bahasa yang mewakili
# ORDER BY 1 maksudnya supaya diurutkan berdasarkan abjad.
SELECT 
	DISTINCT movie_lang
FROM movies
ORDER BY 1


# IN is true when there is a list 
SELECT * FROM movies 
WHERE 
	movie_lang IN ('English', 'Chinese', 'Japanese')
ORDER BY movie_lang 

SELECT * FROM movies 
WHERE 
	movie_lang = 'English'
	OR movie_lang = 'Chinese'
	OR movie_lang = 'Japanese'
ORDER BY movie_lang 

```
