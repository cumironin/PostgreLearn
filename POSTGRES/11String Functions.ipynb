```python
# UPPER Function
# Make string all UPPERCASE

SELECT UPPER ('amazing postgresql')

upper             |
------------------+
AMAZING POSTGRESQL|

# Make select column to directors table on UPPERCASE
SELECT UPPER (first_name) AS first_name, 
	   UPPER(last_name) AS last_name 
FROM directors

first_name|last_name               |
----------+------------------------+
TOMAS     |ALFREDSON               |
PAUL      |ANDERSON                |
WES       |ANDERSON                |

# Make select column to directors table on LOWERCASE
SELECT LOWER (first_name) AS first_name, 
	   LOWER(last_name) AS last_name 
FROM directors

first_name|last_name               |
----------+------------------------+
tomas     |alfredson               |
paul      |anderson                |
wes       |anderson                |


# Make character on Camel Case
SELECT INITCAP('the world is our hand') 

initcap              |
---------------------+
The World Is Our Hand|


# Concat and then make Camel Case : INITCAP on one column
SELECT INITCAP(CONCAT(first_name, ' ', last_name)) AS full_name
FROM directors 
ORDER BY first_name 

full_name                        |
---------------------------------+
Ang Lee                          |
Bruce Lee                        |
Chan-Wook Park                   |
Fernando Meirelles               |
```

```python
# LEFT & RIGHT FUNCTION

# LEFT-----------

# string : character that we want to input, n : number of character that we want to modify
LEFT(string, n)

# Get only 1 character in the left
SELECT LEFT ('ABCD', 1)

left|
----+
A   |

# Get only 2 character in the Right
SELECT LEFT ('ABCD', -2)

left|
----+
AB  |

# Select first name, count and grouping from directors table
SELECT 
	LEFT(first_name ,1) AS initial,
	COUNT(*) AS total_initials
FROM directors 
GROUP BY 1
ORDER BY 1

initial|total_initials|
-------+--------------+
A      |             1|
B      |             1|
C      |             1|
F      |             3|



# RIGHT ----------

# Takeout from right
SELECT RIGHT ('ABCD', 2)

right|
-----+
CD   |


# Takeout from left
SELECT RIGHT ('ABCD', -1)

right|
-----+
BCD  |



```

```python
# SPLIT PART

# Split string into specified string delimiters and returned into nth substring

# Pattern
SPLIT_PART(string, delimiter, position)

# Ambil index ke 3 dari string yang dibatasi dengan koma.
SELECT split_part('ONE, TWO, THREE',',',3) 

split_part|
----------+
 THREE    |
    
# Ambil movie_name, release_date column kemudia ambil dari release_date tahunya saja lalu dinamakan column barunya release_year
SELECT movie_name, release_date, SPLIT_PART(release_date::text, '-', 1) AS release_year
FROM movies

movie_name                                |release_date|release_year|
------------------------------------------+------------+------------+
A Clockwork Orange                        |  1972-02-02|1972        |
Apocalypse Now                            |  1979-08-15|1979        |
Battle Royale                             |  2001-01-04|2001        |


```

```python
# TRIM : remove the longest string that contains a specific character from a string
# LTRIM : remove all character, space by default, from a beginning of a string
# RTRIM : remove all character, space by default, from a end of a string
# BTRIM : is combined between LTRIM() and RTRIM() function

TRIM([LEADING | TRAILING | BOTH] [CHARACTERS] FROM string)

TRIM(LEADING FROM string)

TRIM(TRAILING FROM string)

TRIM(BOTH FROM string)

LTRIM(string, [CHARACTER])
RTRIM(string, [CHARACTER])
BTRIM(string, [CHARACTER])

# remove trailing space Left, Right, or both
SELECT LTRIM('            Amazing Postgresql')

SELECT RTRIM('AMAZING Postgresql          ')

SELECT BTRIM(       'Amazing Postgres'      )

# Remove automatic with TRIM
SELECT TRIM (LEADING FROM '     amazing spiderman'),
       TRIM (TRAILING FROM 'Amazing Spiderman    '),
       TRIM ('   Amazing Spiderman      ')
        
ltrim            |rtrim            |btrim            |
-----------------+-----------------+-----------------+
amazing spiderman|Amazing Spiderman|Amazing Spiderman|
```

```python
# LPAD : functions pad a string on the lefst of specified length with sequence of character 
# RPAD : functions pad a string on the right to a specified length with a sequence of character

# Pattern
LPAD(string, length[, fill])
RPAD(string, length[, fill])

# Pad string on left with  = * and charater all count as 15
SELECT LPAD('Database', 15, '*')

lpad           |
---------------+
*******Database|

# Pad string on right with  = * and charater all count as 15
SELECT RPAD('Database', 15, '*')
rpad           |
---------------+
Database*******|


# Sample

SELECT 
	mv.movie_name, 
	r.revenues_domestic,
#     CAST = conversion to character | TRUNC = Pembulatan pecahan
	LPAD('*', CAST(TRUNC(r.revenues_domestic/10) AS INT), '*') chart
FROM movies mv 
INNER JOIN movies_revenues r ON r.movie_id = mv.movie_id 
ORDER BY 3 DESC
NULLS LAST

movie_name                                |revenues_domestic|chart                                                            |
------------------------------------------+-----------------+-----------------------------------------------------------------+
Titanic                                   |           659.20|*****************************************************************|
Star Wars: A New Hope                     |           461.20|**********************************************                   |
Spider-Man                                |           404.10|****************************************                         |
Spider-Man 2                              |           374.10|*************************************                            |
Forrest Gump                              |           330.30|*********************************                                |

```

```python
# POSITION Function

# Postgresql POSITION() function returns location a substring in a string



# POSITION (substring IN string)
# Check the position first character
SELECT POSITION ('z' IN 'amazing postgre')

position|
--------+
       4|

SELECT POSITION ('is' IN 'This is a computer')

position|
--------+
       3|
    
    
# Check the words position last_name column
SELECT first_name, POSITION ('d' IN last_name) AS last_name_postion FROM directors

first_name|last_name_postion|
----------+-----------------+
Tomas     |                6|
Paul      |                3|
Wes       |                3|

    
```
