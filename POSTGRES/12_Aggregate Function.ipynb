```python
# https://www.w3resource.com/PostgreSQL/postgresql-count-function.php
# https://hevodata.com/learn/postgresql-count/

# COUNT : 
# The PostgreSQL COUNT function counts a number of rows or non-NULL values against a specific column from a table
# When an asterisk(*) is used with count function the total number of rows returns.
# Intinya hitung semua column yang di select & filter

# Syntax:
COUNT (* | [DISTINCT] ALL | column_name)


# Name             Description
# ------------------------------------------------------------------
# column_name      Name of the column
# The asterisk(*)  indicates all the rows.
# DISTINCT         This clause is optional. It indicates uniqueness.
# ALL              This clause is optional. It is default clause.


# Get all row actos and count it
SELECT COUNT(*)  FROM actors 

count|
-----+
  294|
    
```

```python
# https://www.w3resource.com/PostgreSQL/postgresql-max-function.php

# MAX : returns the maximum value, specified by expression in a set of aggregated rows
#       This function accepts an expression including any numeric, string, date, 
#       or time data type values and returns the maximum as a value of the same data type as specified in the expression
# Intinya liat dari nilai terbesar pada sebuah column


# Syntax
SUM (* | [DISTINCT] ALL | column_name)


# Name               Description
# --------------------------------------------------------------------
# column_name        Name of the column
# *                  The asterisk(*) indicates all the rows.
# DISTINCT           This clause is optional. It indicates uniqueness.
# ALL                This clause is optional. It is default clause.

# Cari maximal nilai seluruh di kolom salary
SELECT MAX(salary)  AS "Maximum Salary"
FROM employee;

# Hilangkan double dari data di kolom ini & cari maximal nilai seluruh di kolom salary
SELECT MAX(DISTINCT salary)  AS "Maximum Salary"
FROM employee;

# Dua column
SELECT MAX(salary) "Maximum Salary",
MAX(commission) "Maximum Commission" 
FROM employee;

SELECT MAX(salary)AS "Maximum Salary",
MAX(deduction) AS "Maximum Deduction"
FROM employee
WHERE designame='CLERCK';


# Pilih job_id column, hitung salary berdasarkan group yang sama di job_id
SELECT job_id, MAX(salary) AS "Maximum Salary" 
FROM employees 
GROUP BY job_id;

# Check dengan multiple row
   job_id   | Maximum Salary
------------+----------------
 AC_ACCOUNT |        8300.00
 ST_MAN     |        8200.00
 IT_PROG    |        9000.00
 SA_MAN     |       14000.00
 AD_PRES    |       24000.00
 AC_MGR     |       12000.00
```

```python
# GREATEST : The GREATEST  function returns the “greatest” or “largest” value from the list of expressions.

# Syntax 
# value_1 is the first expression or value.
# value_n is one or more values (separated by comma)
GREATEST ( value_1, [value_n] )


# note : 
> (from the manual: "maximum value of *expression* across all input values")
> greatest(*value* [, ...]) is not an aggregate function
> (from the manual: "The GREATEST function selects the largest value from a list of any number of expressions.")
   
> So max takes the maximum of values coming from several rows, while greatest
> takes the maximum of values coming from 1 row.
   

# sample_1    
select greatest(1,2,greatest(3,4,5));

greatest 

----------

5

# sample_2   
select sum(greatest(1,23,4,5));

 sum 

 -----

  23
    
    

SELECT movie_id, revenues_domestic, revenues_international, GREATEST (revenues_domestic, revenues_international)
FROM movies_revenues

movie_id|revenues_domestic|revenues_international|greatest|
--------+-----------------+----------------------+--------+
      45|            22.20|                  1.30|   22.20| <-- greatest : only one row that evaluate
      13|           199.40|                201.20|  201.20|
      23|           102.10|                      |  102.10|
      44|           158.70|                      |  158.70|
       1|            27.10|                      |   27.10|
```

```python
# https://www.postgresqltutorial.com/postgresql-aggregate-functions/postgresql-avg-function/

# AVG Function

# The AVG() function allows you to calculate the average value of a set.



# example_1

select * from COMPANY;

 id | name  | age | address   | salary
----+-------+-----+-----------+--------
  1 | Paul  |  32 | California|  20000
  2 | Allen |  25 | Texas     |  15000
  3 | Teddy |  23 | Norway    |  20000
  4 | Mark  |  25 | Rich-Mond |  65000
  5 | David |  27 | Texas     |  85000
  6 | Kim   |  22 | South-Hall|  45000
  7 | James |  24 | Houston   |  10000

SELECT AVG(SALARY) FROM COMPANY;

avg
------------------
37142.8571428571


SELECT name, AVG(SALARY) FROM COMPANY GROUP BY name;

 name   |  avg
 -------+-------
  Teddy | 20000
  Paul  | 20000
  Mark  | 65000
  David | 85000
  Allen | 15000
  Kim   | 45000
  James | 10000



# More readable 
SELECT AVG(amount)::numeric(10,2) 
FROM payment;

 avg
------
 4.20

# No Duplicate row column
SELECT AVG(DISTINCT amount)::numeric(10,2)
FROM payment;

 avg
------
 4.20
```
