```python
# Convert Integer to String 
SELECT to_char(100870, '9,99999') 

to_char |
--------+
 1,00870|


# Convert Time To Char
SELECT 
	release_date, 
	to_char(release_date, 'DD-MM-YYYY'),
	to_char(release_date, 'Dy, MM, YYYY')
FROM movies

release_date|to_char   |to_char      |
------------+----------+-------------+
  1972-02-02|02-02-1972|Wed, 02, 1972|
  1979-08-15|15-08-1979|Wed, 08, 1979|
  2001-01-04|04-01-2001|Thu, 01, 2001|


# Convert Numeric to Char
SELECT 
 	movie_id, 
 	revenues_domestic,
 	TO_CHAR(revenues_domestic, '$99999D99')
FROM movies_revenues 

movie_id|revenues_domestic|to_char   |
--------+-----------------+----------+
      45|            22.20|$    22.20|
      13|           199.40|$   199.40|
      23|           102.10|$   102.10|


--------------------------------------------------------------------------------------------------------------------

# SELECT CHARACTER TO NUMBER

SELECT to_number(
	'1428.89',
	'9999.'
) 

to_number|
---------+
     1428|
    

# convert simbol character to number
SELECT to_number(
	'$1,430.78-',
	'L9G999'
) 

to_number|
---------+
     1430|
    
# Add D99 in the back will become (.) point in right
SELECT to_number(
	'$1,430.78-',
	'L9G999D99'
) 

to_number|
---------+
  1430.78|


--------------------------------------------------------------------------------------------------------------------

# TO_DATE

# Convert character to date
SELECT TO_DATE(
	'2020/10/22',
	'YYYY/MM/DD'
)

to_date   |
----------+
2020-10-22|


# Convert character to date (different format)
SELECT TO_DATE(
	'022199',
	'MMDDYY'
)

to_date   |
----------+
1999-02-21|

--------------------------------------------------------------------------------------------------------------------

# CONVERT TO TIMESTAMP

SELECT to_timestamp(
	'2020-10-28 10:30:25',
	'YYYY-MM-DD HH:MI:SS'
); 

to_timestamp                 |
-----------------------------+
2020-10-28 10:30:25.000 +0700|
        
        
SELECT to_timestamp(
	'2020   MAY', 'YYYY MON'
); 

to_timestamp                 |
-----------------------------+
2020-05-01 00:00:00.000 +0700|
        
        
SELECT to_timestamp(
	'2020-01-01 23:8:00', 'YYYY-MM-DD HH24:MI:SS'
); 

to_timestamp                 |
-----------------------------+
2020-01-01 23:08:00.000 +0700|
        

# Note : Check notation above on formating character 
        
```
